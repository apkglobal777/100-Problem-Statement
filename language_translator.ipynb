```python
from torch.nn.modules import padding
#create language translation python application using transformers
from transformers import MarianMTModel, MarianTokenizer
#to select the model
model_name = 'Helsinki-NLP/opus-mt-en-hi'
#load the model
model = MarianMTModel.from_pretrained(model_name)
#load the tokenizer
tokenizer = MarianTokenizer.from_pretrained(model_name)
#to get the input text from user
prompt = input("Enter the text to translate")
#encode the text input using tokenizer
input_tokens = tokenizer(prompt, return_tensors='pt', padding=True)
#pass the token into model and process the data for result
output_token = model.generate(**input_tokens)
#decode the tokens
result = tokenizer.decode(output_token[0], skip_special_tokens=True)
print(result)
```
